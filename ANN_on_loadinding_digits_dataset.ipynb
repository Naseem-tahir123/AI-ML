```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# load dataset
dataset = load_digits()


# preprocessing

scaler = StandardScaler()
X_scaled = scaler.fit_transform(dataset.data)  #centering the data around zero
y = dataset.target

# one-hot-encode the target labels
y = tf.keras.utils.to_categorical(y)


# Splitting the dataset

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)

#ANN - initializing
ann = tf.keras.models.Sequential()

# hidden layers
ann.add(tf.keras.layers.Dense(units=20, activation='relu'))
ann.add(tf.keras.layers.Dense(units=40, activation='relu'))
#output layer
#as target value is binary - AF
ann.add(tf.keras.layers.Dense(units=10, activation='softmax'))

#compiling
#loss - target is binary
ann.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

#training set
ann.fit(X_train, y_train, batch_size = 32, epochs = 50)

# Predict the test set results
y_pred_prob = ann.predict(X_test)

# Convert probabilities to class labels
y_pred = np.argmax(y_pred_prob, axis=1)
y_test_labels = np.argmax(y_test, axis=1)

# Evaluate the model
cm = confusion_matrix(y_test_labels, y_pred)
print(cm)
print("Accuracy:", accuracy_score(y_test_labels, y_pred))
```
