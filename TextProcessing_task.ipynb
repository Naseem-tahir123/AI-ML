# **Basic Feature Extraction**

```python
import pandas as pd

train = pd.read_csv('/content/twcs.csv', nrows=500)
train.head()
```

```python
train.shape
```

# **Number of Words**

```python
train['word_count'] = train['text'].apply(lambda x : len(str(x).split(" ")))

print("\n\n word Count \n\n")
print(train[['text', 'word_count']].head())
```

# **Number of characters**

```python
# Number of characters
print("\n\n Number of Characters \n\n")
train['Char_count'] = train['text'].str.len()
print(train[['text','Char_count']].head())
```

# **Average word Length**

```python
def avg_word_len(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))

train['avg_word_len'] = train['text'].apply(lambda x: avg_word_len(x))
train[['text', 'avg_word_len']].head()
```

# **Number of stopwords**

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

train['stopwords'] = train['text'].apply(lambda x: len([x for x in x.split() if x in stop]))
train[['text', 'stopwords']].head()
```

# **Number Of Special Characters**

```python
train['hashtags'] = train['text'].apply(lambda x: len([x for x in x.split() if x.startswith("#")]))
train[['text', 'hashtags']].head()
```

# **Number of numerics**

```python
train['numerics'] = train['text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
train[['text', 'numerics']].head()
```

# **Number of Uppercase words**

```python
train['upper'] = train['text'].apply(lambda x : len([x for x in x.split() if x.isupper()]))
train[['text', 'upper']].head()
```

# **Basic Pre-processing**

# **1) Lower Case**

```python
train['text'] = train['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
train['text'].head()
```

# **Removing Punctuation**

```python
train['text'] = train['text'].str.replace('[^\w\s]','')
train['text'].head()
```

# **Removal of Stop Words**

```python
train['text'] = train['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop) )
train['text'].head()
```

# **Common word removal**

```python
freq = pd.Series(' '.join(train['text']).split()).value_counts()[:10]
freq
```

```python
# Now, lets remove these words as their presence will not of any use in classification of our text data.
freq = list(freq.index)
train['text'] = train['text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
train['text'].head()
```

# **Rare words removal**

```python
freq = pd.Series(' '.join(train['text']).split()).value_counts()[-10:]
freq
```

# **Spelling correction**

```python
from textblob import TextBlob
train['text'][:5].apply(lambda x: str(TextBlob(x).correct()))
```

# **Stemming**

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
train['text'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

# **Lemmatization**

```python
# most effective option than stemming because it converts the word into its root words
# we usually prefer lemmatization over stemming
from textblob import Word
train['text'] = train['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
train['text'].head()
```
